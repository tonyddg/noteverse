```python
import matplotlib.pyplot as plt
import numpy as np

lims = 5
pts = 5
times = 8

mat = np.array([[1, -2], [-2, 1]])


res_pt = np.zeros((2, pts))

eigenvalue, featurevector = np.linalg.eig(mat)

fig, axe = plt.subplots()

axe.axline((0, 0), (featurevector[0][0], featurevector[1][0]), color = "red")
axe.axline((0, 0), (featurevector[0][1], featurevector[1][1]), color = "yellow")

for angl in np.linspace(-np.pi, np.pi, times + 1):

    ogn_pt = np.zeros((2, 1))
    ogn_pt[0][0] = np.sin(angl)
    ogn_pt[1][0] = np.cos(angl)

    for i in range(0, pts, 1):
        res_pt[0][i] = ogn_pt[0]
        res_pt[1][i] = ogn_pt[1]
        ogn_pt = mat @ ogn_pt

    axe.plot(res_pt[0], res_pt[1])
    axe.scatter(res_pt[0], res_pt[1], 20, np.arange(0, pts, 1) * 10)

axe.axhline(0, -lims, lims, color = "black")
axe.axvline(0, -lims, lims, color = "black")
axe.set_xlim(-lims, lims)
axe.set_ylim(-lims, lims)
axe.set_aspect(1)


# res = featurevector @ np.diag(eigenvalue) @ np.linalg.inv(featurevector)
# print(res)

```

```python
import matplotlib.pyplot as plt
import numpy as np

lims = 3

mat = np.array([[1, -2], [-2, 1]])
base = np.eye(2)

eigenvalue, featurevector = np.linalg.eig(mat)

fig, axes = plt.subplots(2)

rbase = mat @ base 

axes[0].arrow(0, 0, rbase[0][0], rbase[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "black", fc = "black")
axes[0].arrow(0, 0, rbase[0][1], rbase[1][1], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "black", fc = "black")

axes[1].arrow(0, 0, rbase[0][1] + rbase[0][0], rbase[1][1] + rbase[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "black", fc = "black")

axes[0].arrow(0, 0, base[0][0], base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "green", fc = "green")
axes[0].arrow(0, 0, base[0][1], base[1][1], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "green", fc = "green")

axes[1].arrow(0, 0, base[0][1] + base[0][0], base[1][1] + base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "green", fc = "green")


base = np.linalg.inv(featurevector) @ base 

axes[0].arrow(0, 0, base[0][0], base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "blue", fc = "blue")
axes[0].arrow(0, 0, base[0][1], base[1][1], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "blue", fc = "blue")

axes[1].arrow(0, 0, base[0][1] + base[0][0], base[1][1] + base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "blue", fc = "blue")

base = np.diag(eigenvalue) @ base

axes[0].arrow(0, 0, base[0][0], base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "yellow", fc = "yellow")
axes[0].arrow(0, 0, base[0][1], base[1][1], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "yellow", fc = "yellow")

axes[1].arrow(0, 0, base[0][1] + base[0][0], base[1][1] + base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "yellow", fc = "yellow")

base =  featurevector @ base

axes[0].arrow(0, 0, base[0][0], base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "red", fc = "red")
axes[0].arrow(0, 0, base[0][1], base[1][1], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "red", fc = "red")

axes[1].arrow(0, 0, base[0][1] + base[0][0], base[1][1] + base[1][0], length_includes_head=True, head_width=0.05 ,head_length=0.1, ec = "red", fc = "red")

axes[0].set_xlim(-lims, lims)
axes[0].set_ylim(-lims, lims)
axes[0].set_aspect(1)

axes[1].set_xlim(-lims, lims)
axes[1].set_ylim(-lims, lims)
axes[1].set_aspect(1)

print(np.diag(eigenvalue))
print(featurevector)
print(np.linalg.inv(featurevector))
```
